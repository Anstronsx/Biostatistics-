#Correlación

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = {
    'T0': [4.1, 4.3],
    'T1': [3.8, 3.6],
    'T2': [3.7, 4.0],
    'T3': [4.0, 4.1]
}

tratamientos = ['PA vs VP', 'PA vs VV']

df = pd.DataFrame(data, index=tratamientos)

# Calcutor of correlation
correlation_matrix = df.corr()

# plt
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True, square=True, fmt='.2f')

#plt graph

plt.title('Correlation')
plt.show()
```

```python
# DataFrame
df_transposed = df.T

# Graph
plt.figure(figsize=(8, 6))
plt.gca().patch.set_facecolor('white')  # Cambiar el color de fondo de la figura
plt.gca().spines['top'].set_visible(False)  # Opcional: Ocultar los bordes del gráfico
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_linewidth(0.5)
plt.gca().spines['bottom'].set_linewidth(0.5)

for tratamiento in df_transposed.columns:
    plt.plot(df_transposed.index, df_transposed[tratamiento], marker='o', label=tratamiento)

# title of the graph in Spanish
plt.title('Actividad Antagónica PA vs VP')
plt.xlabel('Tratamientos')
plt.ylabel('Promedio')
plt.legend(title='Correlación')
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
# DAtaframe copy
df_transposed = df.T.copy()

# boxplot
plt.figure(figsize=(8, 5))
plt.boxplot([df_transposed[tratamiento] for tratamiento in df_transposed.columns],
            labels=df_transposed.columns,
            widths=0.08)


#  eje y
plt.ylim(bottom=3)

# Title
plt.title('Actividad Antagónica PA vs VP')
plt.xlabel('Tratamientos')
plt.ylabel('Promedio')
plt.grid(True, linestyle='-', alpha=1)


#plt
plt.show()
```

```python
# Creando los violin plots
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_transposed)
plt.title('Violin Plots de los Tratamientos')
plt.ylabel('Valor')
plt.xlabel('Tratamiento')
plt.show()
```

```python
def scatter_plot_custom(df, colors):
    g = sns.PairGrid(df)
    for i, col in enumerate(df.columns):
        g.map_diag(sns.histplot, color=colors[i])
        g.map_offdiag(sns.scatterplot, color=colors[i])
    plt.suptitle('Dispersion entre los Probióticos vs Patogenos', y=1.02)
    plt.show()

#colours
colors = ['blue', 'red', 'yellow', 'skyblue']

scatter_plot_custom(df, colors)
```

```python
# Interactive scatter plot
fig = px.scatter_matrix(df, dimensions=df.columns, title="Matriz de Dispersión Interactiva")
fig.show()

# Heatmap
fig = go.Figure(data=go.Heatmap(
                   z=df.corr().values,
                   x=df.columns,
                   y=df.columns,
                   colorscale='Viridis'))
fig.update_layout(title='Matriz de Correlación Interactiva')
fig.show()
```

```python
# Normalising the data for the radar chart
df_normalized = (df - df.min()) / (df.max() - df.min())

# plt
fig = px.line_polar(df_normalized, r=df_normalized.mean(), theta=df.columns, line_close=True, title='Tratamientos')
fig.show()
```

```python
plt.figure(figsize=(10, 6))
for column in df.columns:
    sns.kdeplot(df[column], shade=True, label=column)
plt.title('Gráficos de Densidad de los Tratamientos')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```
